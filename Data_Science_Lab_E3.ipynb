```python
mriirs = '/content/drive/MyDrive/Data Science_Lab/BMW Sales Data (2010-2024)_E3.csv'
sayem = pd.read_csv(mriirs)
```

```python
# View first 5 rows to check what data looks like
print("First 5 rows of the dataset:")
display(sayem.head())
```

```python
# Check the basic shape (rows Ã— columns)
print("\n Dataset Shape (rows, columns):", sayem.shape)
```

```python
# Get information about columns, datatypes, and missing values
print("\n Dataset Information:")
sayem.info()
```

```python
# Show summary statistics for numerical columns
print("\n Statistical Summary:")
display(sayem.describe())
```

```python
# Show the column names
print("\n Column Names:")
print(sayem.columns.tolist())
```

```python
# Rename columns (if names are long or unclear)
# Example: rename 'Sales Volume' to 'Sales_Volume'
sayem.rename(columns={'Sales Volume': 'Sales_Volume'}, inplace=True)
print("\n Columns after renaming:")
print(sayem.columns.tolist())
```

```python
# Check for missing values in each column
print("\n Missing values per column:")
print(sayem.isnull().sum())
```

```python
#Fill missing values with mean (for numbers) or mode (for text)
for col in sayem.columns:
    if sayem[col].dtype == 'object':                   # if column is text
        sayem[col].fillna(sayem[col].mode()[0], inplace=True)
    else:                                              # if column is numeric
        sayem[col].fillna(sayem[col].mean(), inplace=True)

print("\n Missing values after filling:")
print(sayem.isnull().sum())
```

```python
# Sorting the dataset
# Sort by 'Year' in ascending order
sayem_sorted = sayem.sort_values(by='Year', ascending=True)
print("\n Dataset sorted by Year:")
display(sayem_sorted.head())
```

```python
# Filtering data (select only rows where Price > 50000)
print("\n Cars with Price > 50000:")
display(sayem[sayem['Price_USD'] > 50000].head())
```

```python
# Selecting specific columns (Model, Year, Price)
print("\n Selecting few important columns:")
display(sayem[['Model', 'Year', 'Price_USD']].head())
```

```python
# Grouping data: Find average price per year
print("\n Average Price per Year:")
avg_price = sayem.groupby('Year')['Price_USD'].mean()
display(avg_price)
```

```python
# Find the most popular BMW model (highest sales volume)
if 'Sales_Volume' in sayem.columns:
    print("\n Most Sold BMW Model:")
    top_model = sayem.groupby('Model')['Sales_Volume'].sum().sort_values(ascending=False)
    display(top_model.head(5))
```

```python
# Check all column names to confirm spelling
print("Available columns in dataset:")
print(sayem.columns.tolist())
```

```python
# Create a new derived column (example)
# Price_per_KM = Price_USD / Mileage_KM
if {'Price_USD', 'Mileage_KM'}.issubset(sayem.columns):
    # Create a new column named 'Price_per_KM'
    sayem['Price_per_KM'] = sayem['Price_USD'] / sayem['Mileage_KM']

    print("\n New column added: Price_per_KM = Price_USD / Mileage_KM")
    # Display the first 5 rows of the new column
    display(sayem[['Model', 'Mileage_KM', 'Price_USD', 'Price_per_KM']].head())
else:
    print(" Columns 'Price_USD' or 'Mileage_KM' not found in the dataset.")
```

```python
# Save cleaned and updated dataset to Drive
output_path = '/content/drive/MyDrive/Data Science_Lab/bmw_sales_cleaned.csv'
sayem.to_csv(output_path, index=False)
print(f"\n Updated dataset saved to: {output_path}")
```
