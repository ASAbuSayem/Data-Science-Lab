```python
# Import necessary libraries
import pandas as pd
import numpy as np
```

```python
file_path = '/content/drive/MyDrive/Data Science_Lab/dermatology_database_E2.csv'
sayem = pd.read_csv(file_path) ## Read the CSV file and store it into a DataFrame named 'sayem'
```

```python
sayem.head()
```

```python
# Display the first 5 rows to understand the structure
print("First 5 rows of the dataset:")
display(sayem.head())
```

```python
# Get general info about dataset
# Shows total rows, columns, datatypes, and missing values
print("\n Dataset Information:")
sayem.info()
```

```python
# Display summary statistics for all numerical columns
# Includes count, mean, std, min, max, etc.
print("\n Statistical Summary of Numerical Columns:")
display(sayem.describe())
```

```python
# Select only numerical columns for computation
num_cols = sayem.select_dtypes(include=np.number).columns
print("\n Numerical Columns:", list(num_cols))
```

```python
# Compute central tendency measures
print("\n Mean values of each numerical column:")
print(sayem[num_cols].mean())
```

```python
print("\n Median values of each numerical column:")
print(sayem[num_cols].median())
```

```python
print("\n Mode values of each numerical column:")
print(sayem[num_cols].mode().iloc[0])   # .iloc[0] takes the first mode row ( mode() returns the most frequent value(s). It may return multiple rows,
# so we use .iloc[0] to pick the first set of mode values.)
```

```python
# Compute data spread: variance & standard deviation
print("\n Variance of numerical columns:")
print(sayem[num_cols].var())
```

```python
print("\n Standard Deviation of numerical columns:")
print(sayem[num_cols].std())
```

```python
# Perform basic mathematical operations
print("\n Adding 10 to all numerical columns:")
display(sayem[num_cols] + 10)

print("\n Subtracting 5 from all numerical columns:")
display(sayem[num_cols] - 5)

print("\n Multiplying all numerical columns by 2:")
display(sayem[num_cols] * 2)

print("\n Dividing all numerical columns by 2:")
display(sayem[num_cols] / 2)
```

```python
# Correlation matrix to find relationships between numeric features
print("\n Correlation Matrix:")
display(sayem[num_cols].corr())
```

```python
# Using NumPy for element-wise operations
# Example: Apply square root to first 5 rows of numeric columns
print("\n Applying sqrt() to first 5 rows of numerical data:")
display(np.sqrt(sayem[num_cols].head()))
```

```python
# Conditional filtering example
# (Adjust column name if needed, e.g., 'age' or 'erythema')
if 'age' in sayem.columns:
    # Convert 'age' column to numeric, coercing errors to NaN and filling NaN with 0
    sayem['age'] = pd.to_numeric(sayem['age'], errors='coerce').fillna(0)
    print("\n Records where age > 40:")
    display(sayem[sayem['age'] > 40].head())
```

```python
# Create a derived column example
# If dataset has numeric columns 'erythema' and 'scaling'
if {'erythema', 'scaling'}.issubset(sayem.columns):
    sayem['sum_symptoms'] = sayem['erythema'] + sayem['scaling']
    print("\n New Derived Column Added: sum_symptoms = erythema + scaling")
    display(sayem[['erythema', 'scaling', 'sum_symptoms']].head())
```

```python
# Save the processed dataset
output_path = '/content/drive/MyDrive/dermatology_numerical_ops.csv'
sayem.to_csv(output_path, index=False)
print(f"\n Updated dataset saved to: {output_path}")
```
