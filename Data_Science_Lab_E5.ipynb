```python
# Dataset: Air Quality Index (Global)
# Author: Md. Abu Sayem
# Import Libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from datetime import datetime
```

```python
# Enable beautiful visual style
sns.set(style="whitegrid", palette="pastel")
```

```python
# Load Dataset from Google Drive
file_path = '/content/drive/MyDrive/Data Science_Lab/data_date.csv'
sayem = pd.read_csv(file_path)
```

```python
# View first few rows
print("First 5 rows of dataset:")
display(sayem.head())
```

```python
# Basic Info & Summary
print("\n Dataset Information:")
sayem.info()
```

```python
print("\n Summary Statistics:")
display(sayem.describe())
```

```python
# Check for missing values
print("\n Missing Values per Column:")
print(sayem.isnull().sum())
```

```python
# Visualize missing values with heatmap
plt.figure(figsize=(8,4))
sns.heatmap(sayem.isnull(), cbar=False, cmap="Reds")
plt.title("Missing Values Heatmap")
plt.show()
```

```python
# Convert date/time column to datetime format
# Replace column name with the actual date column (e.g. 'Date' or 'Datetime')
sayem['Date'] = pd.to_datetime(sayem['Date'], errors='coerce')
```

```python
# Sort dataset by date for time series
sayem.sort_values('Date', inplace=True)
```

```python
# Extract Year, Month, and Day (for seasonal analysis)
sayem['Year'] = sayem['Date'].dt.year
sayem['Month'] = sayem['Date'].dt.month_name()
sayem['Day'] = sayem['Date'].dt.day
```

```python
# Drop rows with missing dates
sayem.dropna(subset=['Date'], inplace=True)
```

```python
# Check AQI column name and rename if needed
# Example: If it's 'AQI Value' or 'AirQualityIndex', rename it
sayem.rename(columns={'AQI Value': 'AQI'}, inplace=True)
```

```python
# EXPLORATORY DATA ANALYSIS (EDA)
print("\n Dataset Shape (rows, columns):", sayem.shape)
print("\n Unique Countries:", sayem['Country'].nunique())
```

```python
# Top countries with most records
top_countries = sayem['Country'].value_counts().head(10)
plt.figure(figsize=(16,4))
sns.barplot(x=top_countries.index, y=top_countries.values, palette="coolwarm")
plt.title("Top 10 Countries with Most AQI Records")
plt.ylabel("Number of Records")
plt.xlabel("Country")
plt.show()
```

```python
#TIME SERIES VISUALIZATION
# Average AQI over Time (Line Plot)
plt.figure(figsize=(15,5))
sns.lineplot(x='Date', y='AQI', data=sayem, color='green')
plt.title("Air Quality Index (AQI) Over Time")
plt.xlabel("Date")
plt.ylabel("AQI Level")
plt.show()
```

```python
# Yearly AQI Trend
yearly_aqi = sayem.groupby('Year')['AQI'].mean()
plt.figure(figsize=(8,4))
sns.barplot(x=yearly_aqi.index, y=yearly_aqi.values, palette='mako')
plt.title("Average AQI per Year")
plt.ylabel("Mean AQI")
plt.xlabel("Year")
plt.show()
```

```python
# Monthly Trend (Seasonality)
monthly_aqi = sayem.groupby('Month')['AQI'].mean()
plt.figure(figsize=(16,4))
sns.lineplot(x=monthly_aqi.index, y=monthly_aqi.values, marker='o', color='orange')
plt.title("Average Monthly AQI Trend")
plt.ylabel("AQI")
plt.xlabel("Month")
plt.show()
```

```python
# Boxplot to show AQI variation by Country
plt.figure(figsize=(65,15))
sns.boxplot(x='Country', y='AQI', data=sayem, palette='Set2')
plt.title("AQI Distribution by Country")
plt.xticks(rotation=45)
plt.show()
```

```python
# Correlation Heatmap (if multiple numeric columns exist)
plt.figure(figsize=(8,4))
sns.heatmap(sayem.corr(numeric_only=True), annot=True, cmap='coolwarm')
plt.title("Correlation Heatmap of AQI Dataset")
plt.show()
```

```python
# Distribution of AQI
plt.figure(figsize=(8,4))
sns.histplot(sayem['AQI'], bins=30, kde=True, color='teal')
plt.title("Distribution of AQI Values")
plt.xlabel("Air Quality Index (AQI)")
plt.ylabel("Frequency")
plt.show()
```
