```python
# Check for missing values
# sayem.isna() â€” returns a Boolean DataFramesayem) of the same size as df, where True indicates missing (NaN) values.
#.sum() counts the number of missing values per column, helping identify columns needing cleaning.
sayem.isna().sum()
```

```python
# Save a cleaned version back to Drive
cleaned_path = '/content/drive/MyDrive/health_fitness_cleaned.csv'
sayem.to_csv(cleaned_path, index=False)  # index=False to avoid writing the DataFrameâ€™s(sayem) index as a separate column.
```

```python
# Visualization
plt.figure(figsize=(15,8))
sns.heatmap(sayem.corr(numeric_only=True), annot=True, cmap="coolwarm")
plt.title("Correlation Heatmap")
plt.show()
```

```python
# Histogram for numerical columns
sayem.hist(figsize=(20, 15), bins=20, color='skyblue')
plt.suptitle("Distribution of Numerical Features", fontsize=12)
plt.show()
```

```python
# Bar chart example: Distribution of gender
plt.figure(figsize=(6, 4))
sns.countplot(data=sayem, x='gender')
plt.title('Distribution of Gender')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.show()
```

```python
# Pie chart example: Distribution of smoking status
plt.figure(figsize=(6, 6))
sayem['smoking_status'].value_counts().plot(kind='pie', autopct='%1.1f%%', startangle=90)
plt.title('Distribution of Smoking Status')
plt.ylabel('') # Remove the default 'smoking_status' label on the y-axis
plt.show()
```

```python
# Histogram example: Distribution of age
plt.figure(figsize=(8, 6))
sns.histplot(data=sayem, x='age', bins=20, kde=True)
plt.title('Distribution of Age')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```
